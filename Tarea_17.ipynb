__Ejercicio 1:__ ¿Como podemos evaluar funciones en matrices?
__Hint:__ ¿que operaciones sabemos que estan bien definidas sobre matrices?

La operación de evaluar una matriz A en la función $f(x)=Exp(x)$ está bien definida, utilizando la serie de Taylor de f, ya que hacer esto significa obtener una serie cuyos elementos son el producto de A consigo misma varias veces. Así, evaluar f en A está bien definida pues el producto de matrices está bien definido. Por lo anterior, podemos extender dicha operación para evaluar la matriz A en cualquier función $g(x)$ utilizando la serie de Taylor de g.

__Ejercicio 2:__ ¿Como se define una matriz simétrica?

Una matriz cuadrada A es simétrica si y sólo si considerando los elementos de A, $a_{ij}$, se cumple que $a_{ij} = a_{ji}$ para cualquier $i,j$.

__Ejercicio 3:__ ¿Como puede construir una matriz simétrica a partir de cualquier matriz cuadrada?

Si se tiene una matriz A de dimensión n, se pueden obtener sus n eigenvalores, digamos $\lambda _1, \lambda _2, ... , \lambda _n$. Con éstos se puede formar una matriz diagonal, D, de dimensión n cuyas entradas sean los n eigenvalores de A. Además, se puede obtener una matriz, V, tal que sus columnas estén formadas por los eigenvectores de A (éstos se pueden obtener con los eigenvalores $\lambda _i$). Con esto, se puede definir la matriz $C = V*D*V^T - A$, donde $V^T$ es la transpuesta de V; tal que C es simétrica.

__Ejercicio 4:__ Construya el propagador $e^{A t}$ para cada tiempo sobre una malla y uselo para propagar un estado inicial $\vec x_0$ cualquiera. Utilice:
$$A=\left(\begin{array}{cc} -1 & 1\\ -1 & -1 \end{array} \right).$$
* Grafique la trayectoria en el espacio de fase.

```julia
T=linspace(-1,1,1000)
Z=[-1 1 ; -1 -1]
z0 = [1,1]
Zrx=[]
Zry=[]
for t in T
    vals=eigvals(t*Z)
    w=eigvecs(t*Z)
    d = diagm(exp.(vals))
    Z2=w*d*inv(w)
    Vec=Z2*z0
    push!(Zrx,Vec[1])
    push!(Zry,Vec[2])
end
plt.plot(Zrx,Zry)
plt.xlabel("x(t)")
plt.ylabel("y(t)")
```
