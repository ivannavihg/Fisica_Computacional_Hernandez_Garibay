## Ejercicio 1

```julia
function euler_implicito_pf(f,x0,listh)
    x = x0
    S = []
    h = (last(listh)-listh[1])/length(listh)
    for i in listh
        z = x
        for j in 1:2
            z = x + h*f(z,i+h)
        end
        x = z
        push!(S,x)
    end
    listh, S
end
```

```julia
A1,A2=euler_implicito_pf((x,t)->-2.5x,1,linspace(0,8,40))
#println(A2)
plt.scatter(A1,A2, alpha=0.5)
e1 = []
for i in linspace(0,8,100)
    push!(e1,e^(-2.5*i))
end
plt.plot(linspace(0,8,100),e1, color="green")
```

## Ejercicio 2

```julia
function metodo_euler_medio(f,x0,listh)
    #listh = linspace(t0,tf,(tf-t0)/h)
    x = x0
    S = []
    h = (last(listh)-listh[1])/length(listh)
    for i in listh
        x = x + h*f(x +(h/2.)*f(x,i),i+h/2.)
        push!(S,x)
    end
    listh, S
end
```

```julia
A3,A4=metodo_euler_medio((x,t)->-2.5x,1,linspace(0,8,40))
plt.scatter(A3,A4, alpha=0.5)
eA3 = []
for i in A3
    push!(eA3,e^(-2.5*i))
end
plt.plot(A3,eA3, color="green")
```

## Ejercicio 3

```julia
function metodo_RK4(f,x0,listh)
    #listh = linspace(t0,tf,(tf-t0)/h)
    x = x0
    S = []
    h = (last(listh)-listh[1])/length(listh)
    for i in listh
        k1 = f(x,i)
        k2 = f(x+ (h/2.)*k1, i+h/2.)
        k3 = f(x+ (h/2.)*k2, i+h/2.)
        k4 = f(x + h*k3,i+h)
        x = x + (h/6.)*(k1 + 2.*k2 + 2.*k3 + k4)
        push!(S,x)
    end
    listh, S
end
```

```julia
A5,A6=metodo_RK4((x,t)->-2.5x,1,linspace(0,8,40))
plt.scatter(A5,A6, alpha=0.5)
e1 = []
for i in linspace(0,8,100)
    push!(e1,e^(-2.5*i))
end
plt.plot(linspace(0,8,100),e1, color="green")
```

## Ejercicio 6

Si se considera un oscilador armnico simple, con constante de resorte $k=1$, y con masa $m=1$, con condiciones iniciales $x_0 = 1$ y $v_0 = 0$, la Segunda Ley de Newton es para este caso:

\begin{equation}
\frac{d^2x(t)}{dx^2}  = -kx(t)
\end{equation}

es decir, 

\begin{equation}
x''(t) + kx(t) = 0
\end{equation}

Si $u = x'(t)$, entonces $u'(t) = x''(t)$. Esto significa que la ecuacin original cumple con dos condiciones:
1. $u(t) = x'(t)$
2. $u'(t) = -x(t)$

Entonces se puede resolver con mtodo de Euler o Runge-Kutta de dos dimensiones:

### Euler

```julia
function metodo_euler_2(f,g,CI,listh) #CI es arreglo con condiciones iniciales
    #listh = linspace(t0,tf,(tf-t0)/h)
    x=CI[1]
    u=CI[2]
    Sx = []
    Su = []
    h = (last(listh)-listh[1])/length(listh)
    for i in listh
        x = x + h*g(i,x,u)
        u = u + h*f(i,x,u)
        push!(Sx,x)
        push!(Su,u)
    end
    listh, Sx, Su
end;
```

```julia
A1,A2,A3=metodo_euler_2((i,x,u)-> -x,(i,x,u)-> u,[1,0],linspace(0,2pi,63))
plt.figure(figsize=(10,5))
plt.scatter(A1,A2,alpha=0.5, label="x(t)")
plt.scatter(A1,A3,alpha=0.5, label="u(t)")
plt.xlabel("t")
plt.legend()
```

De esta forma, como la energa total del oscilador armnico es $E = T + U$, y como $T = \frac{u^2}{2}$ y $U = \frac{x^2}{2}$, entonces se grafica la energa:

```julia
T = A3.^2 / 2.
U = A2.^2 / 2.
E = T + U
plt.figure(figsize=(10,5))
plt.scatter(A1,T,alpha=0.5, color="red", label="T")
plt.scatter(A1,U,alpha=0.5, color="blue", label="U")
plt.scatter(A1,E,alpha=0.5, color="orange", label="E")
plt.legend()
plt.xlabel("t")
plt.ylabel("Energa")

plt.figure(figsize=(10,5))
plt.scatter(A2,T,alpha=0.5, color="red", label="T")
plt.scatter(A2,U,alpha=0.5, color="blue", label="U")
plt.scatter(A2,E,alpha=0.5, color="orange", label="E")
plt.legend()
plt.xlabel("x")
plt.ylabel("Energa")
```

### Runge-Kutta

```julia
function metodo_RK4_2(f,g,CI,listh)
    #listh = linspace(t0,tf,(tf-t0)/h)
    x=CI[1]
    u=CI[2]
    Sx = []
    Su = []
    h = (last(listh)-listh[1])/length(listh)
    for i in listh
        m1 = g(x,u,i)
        k1 = f(x,u,i)
        m2 = g(x+ (h/2.)*m1,u+ (h/2.)*k1, i+h/2.)
        k2 = f(x+ (h/2.)*m1,u+ (h/2.)*k1, i+h/2.)
        m3 = g(x+ (h/2.)*m2,u+ (h/2.)*k2, i+h/2.)
        k3 = f(x+ (h/2.)*m2,u+ (h/2.)*k2, i+h/2.)
        m4 = g(x+ h*m3,u+ h*k3, i+h)
        k4 = f(x+ h*m3,u+ h*k3, i+h)
        
        x = x + (h/6.)*(m1 + 2.*m2 + 2.*m3 + m4)
        u = u + (h/6.)*(k1 + 2.*k2 + 2.*k3 + k4)
        push!(Sx,x)
        push!(Su,u)
    end
    listh, Sx, Su
end
```

```julia
R1,R2,R3=metodo_RK4_2((x,u,i)-> -x,(x,u,i)-> u,[1,0],linspace(0,2pi,63))
plt.figure(figsize=(10,5))
plt.scatter(A1,A2,alpha=0.5, label="x(t)")
plt.scatter(A1,A3,alpha=0.5, label="u(t)")
plt.xlabel("t")
plt.legend()
```

```julia
T = R3.^2 / 2.
U = R2.^2 / 2.
E = T + U
plt.figure(figsize=(10,5))
plt.scatter(R1,T,alpha=0.5, color="red", label="T")
plt.scatter(R1,U,alpha=0.5, color="blue", label="U")
plt.scatter(R1,E,alpha=0.5, color="orange", label="E")
plt.legend()
plt.xlabel("t")
plt.ylabel("Energa")

plt.figure(figsize=(10,5))
plt.scatter(R2,T,alpha=0.5, color="red", label="T")
plt.scatter(R2,U,alpha=0.5, color="blue", label="U")
plt.scatter(R2,E,alpha=0.5, color="orange", label="E")
plt.legend()
plt.xlabel("x")
plt.ylabel("Energa")
```
