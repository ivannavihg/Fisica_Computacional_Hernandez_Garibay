```julia
using PyPlot;
plt = PyPlot;
```

```julia
function bifurcaciones(xint,muint)
    long1=length(muint)
    long2=length(xint)
    m=1000
    j=1
    Arx = zeros(long1,long2)
    Armu= zeros(long1,long2)
    for mu in muint
        xint2=xint
        for i in 1:m
        xint2 = 1-mu*xint2.^2
        end
        for k in 1:long2
            Armu[j,k]=mu
        end
        Arx[j,1:end]=xint2
        j = j+1
    end
    Arx,Armu
end
```

```julia
A1,A2=bifurcaciones(linspace(-1,1,200),linspace(0,2,200))
plt.scatter(A2,A1,alpha=0.3,s=5)
plt.ylabel("x0")
plt.xlabel("mu")
```

```julia
function bifurcaciones2(xint,muint)
    long1=length(muint)
    long2=length(xint)
    m=1000
    j=1
    Arx = zeros(long1,long2)
    Armu= zeros(long1,long2)
    for mu in muint
        xint2=xint
        for i in 1:m
        xint2 = mu*xint2.*(1-xint2)
        end
        for k in 1:long2
            Armu[j,k]=mu
        end
        Arx[j,1:end]=xint2
        j = j+1
    end
    Arx,Armu
end
```

```julia
A1,A2=bifurcaciones2(linspace(0,1,200),linspace(0,4,200))
plt.scatter(A2,A1,alpha=0.3,s=5)
plt.ylabel("x0")
plt.xlabel("mu")
```
